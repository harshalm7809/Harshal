```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('ind.csv')

file = file[:-2]

iso = {'Andhra Pradesh':'AP','Arunachal Pradesh':'AR','Assam':'AS','Bihar':'BR','Chhattisgarh':'CT','Goa':'GA','Gujarat':'GJ','Haryana':'HR',
'Himachal Pradesh':'HP','Jammu and Kashmir':'JK','Jharkhand':'JH','Karnataka':'KA','Kerala':'KL','Madhya Pradesh':'MP',
'Maharashtra':'MH','Manipur':'MN','Meghalaya':'ML','Mizoram':'MZ','Nagaland':'NL','Odisha':'OR','Punjab':'PB','Rajasthan':'RJ',
'Sikkim':'SK','Tamil Nadu':'TN','Tripura':'TR','Uttar Pradesh':'UP','Uttarakhand':'UT','West Bengal':'WB',
'Andaman and Nicobar Islands':'AN','Chandigarh':'CH','Dadra and Nagar Haveli and Daman and Diu':'DN','Chhattisgarh':'CT',
'Lakshadweep':'LD','Delhi':'DL','Puducherry':'PY','Ladakh':'LA','Telangana':'TE',}

file["States"] = file.replace({"States":iso})
#print(file["States"])

```

```python
fig=plt.figure(figsize=(16, 8))
plt.bar(x=np.arange(1,36),height=file['Active'], width=0.7,color = 'red')
plt.title("Active cases In India (12/7/2020)")
plt.xticks(np.arange(1,36), file['States'], rotation=90, fontsize=14)

plt.xlabel("States")
plt.ylabel("Active Cases")
plt.grid(color='blue')
plt.show()

    
    
```

```python
fig=plt.figure(figsize=(16, 8))
plt.bar(x=np.arange(1,36),height=file['Recovered'], width=0.7,color = 'Green')
plt.title("Recovered cases In India (12/7/2020)")
plt.xticks(np.arange(1,36), file['States'], rotation=90, fontsize=14)

plt.xlabel("States")
plt.ylabel("Recovered Cases")
plt.grid(color='blue')
plt.show()
```
